```python
%matplotlib inline
import numpy as np
import scipy.stats as stats
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

plt.plot([-1, 0, 1], [0.5, 0.5, 0.5])
plt.axis([-2, 2, 0, 1.2])

plt.plot([-2, -1, 0, 1, 2], [0, 0, 0.5, 1, 1])

plt.show()
```

```python
xs = np.linspace(0,1, 20)
ys = [4*np.power(num, 3) for num in xs]

plt.plot(xs, ys)
plt.axis([0, 1, 0, 4])

yz = [np.power(num, 4) for num in xs]
plt.plot(xs, yz)

plt.show()
```

```python
xs = np.linspace(0,4, 20)
ys = [3/16. * np.sqrt(num) for num in xs]

plt.plot(xs, ys)
#plt.axis([0, 1, 0, 4])

yz = [np.power(num, 1.5) / 8. for num in xs]
plt.plot(xs, yz)

plt.show()
```

```python
xs = np.linspace(0.01,1)
ys = [1/(4 * np.power(num, .75)) for num in xs]

plt.plot(xs, ys)
#plt.axis([0, 1, 0, 4])

yz = [np.power(num, 0.25) for num in xs]
plt.plot(xs, yz)

plt.show()
```
