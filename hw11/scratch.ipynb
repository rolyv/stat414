```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = [1, 1, 1, 1 ,2, 2, 2, 2]
y = [1, 2, 3, 4, 1, 2, 3, 4]

plt.axis([0,3, 0, 5])
plt.scatter(x,y)

labels = ['{0}'.format((X+Y)/32.) for X in range(1,3) for Y in range(1,5)]
for label, x, y in zip(labels, x, y):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (35, -5),
        textcoords = 'offset points', ha = 'center', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5))

Y_labels = ['{0}'.format(sum([(X+Y)/32. for X in range(1,3)])) for Y in range(1,5)]
for label, y in zip(Y_labels, range(1,5)):
    plt.annotate(label, xy=(0, y), xytext=(-20, -5),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5))
    
X_labels = ['{0}'.format(sum([(X+Y)/32. for Y in range(1,5)])) for X in range(1,3)]
for label, x in zip(X_labels, range(1,3)):
    plt.annotate(label, xy=(x, 0), xytext=(20, -30),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5))
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = [1, 1, 1, 1 ,2, 2, 2, 2]
y = [1, 2, 3, 4, 1, 2, 3, 4]

fig, ax = plt.subplots(figsize=(8,8))
ax.axis([0,4,0,5])
ax.scatter(x,y)

ax.plot([0,3.5],[1.5, 1.5], 'b')
ax.plot([0,3.5],[2.5, 2.5], 'b')
ax.plot([0,3.5],[3.5, 3.5], 'b')
ax.plot([3.0,3.0], [0,5], 'b')

labels = ['{0:.4f}'.format((X+Y)/(2. * Y + 3)) for X in range(1,3) for Y in range(1,5)]
for label, x, y in zip(labels, x, y):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (35, -5),
        textcoords = 'offset points', ha = 'center', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5))
    
cond_pmf_labels = ['g(x|{0})'.format(y) for y in range(1,5)]
for label, y in zip(cond_pmf_labels,range(1,5)):
    plt.annotate(
        label,
        xy=(3.0, y), xytext=(25,-5),
        textcoords = 'offset points', ha = 'center', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5))

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = [1, 1, 1, 1 ,2, 2, 2, 2]
y = [1, 2, 3, 4, 1, 2, 3, 4]

fig, ax = plt.subplots(figsize=(8,8))
ax.axis([0,4,0,5])
ax.scatter(x,y)

ax.plot([1,2.5],[4.25, 4.25], 'b')
ax.plot([1.8,1.8], [0,5], 'b')

labels = ['{0:.4f}'.format((X+Y)/(4. * X + 10)) for X in range(1,3) for Y in range(1,5)]
for label, x, y in zip(labels, x, y):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (35, -5),
        textcoords = 'offset points', ha = 'center', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5))
    
cond_pmf_labels = ['h(y|{0})'.format(x) for x in range(1,3)]
for label, x in zip(cond_pmf_labels,range(1,3)):
    plt.annotate(
        label,
        xy=(x, 4.3), xytext=(30,15),
        textcoords = 'offset points', ha = 'center', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5))

plt.show()
```
