```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

mean = 50
variance = 36
sigma = math.sqrt(variance)
x = np.linspace(-20+mean,20+mean,100)
X, = plt.plot(x,mlab.normpdf(x,mean,sigma), label='X')

variance = 36/9.
sigma = math.sqrt(variance)
sample_9, = plt.plot(x,mlab.normpdf(x,mean,sigma), label='Sample n=9')

variance = 36/36.
sigma = math.sqrt(variance)
sample_36, = plt.plot(x,mlab.normpdf(x,mean,sigma), label='Sample n=36')

plt.legend()

plt.show()
```
